```python
import tkinter as tk

# Constantes

LARGEUR = 600
HAUTEUR = 400


# Fonctions

def creer_balle():
    """Dessine un rond bleu et retourne son identifiant
     et les valeurs de déplacements dans une liste"""
    x, y = LARGEUR // 2, HAUTEUR // 2
    dx, dy = 3, 5
    rayon = 20
    cercle = canvas.create_oval((x-rayon, y-rayon),
                                (x+rayon, y+rayon),
                                fill="blue")
    return [cercle, dx, dy]


def mouvement():
    """Déplace la balle et ré-appelle la fonction avec un compte-à-rebours"""
    rebond()
    canvas.move(balle[0], balle[1], balle[2])
    canvas.after(20, mouvement)


def rebond():
    """Fait rebondir la balle sur les bords du canevas"""
    global balle
    x0, y0, x1, y1 = canvas.coords(balle[0])
    if x0 <= 0 :
        balle[1] = -balle[1]
        canvas.itemconfigure(balle[0], fill="white")
    if x1 >= 600:
        balle[1] = -balle[1]
        canvas.itemconfigure(balle[0], fill="red")       
    if y0 <= 0 :
        balle[2] = -balle[2]
        canvas.itemconfigure(balle[0], fill="white")
    if y1 >= 400:
        balle[2] = -balle[2]
        canvas.itemconfigure(balle[0], fill="red")
       
#def creer_canvas():
    #balle =creer_balle()


# programme principal

racine = tk.Tk()
canvas = tk.Canvas(racine, bg="black", width=600, height=400)
#canvas.bind('<Button-2>',creer_canvas )
ligne =canvas.create_line(0,600,0,0,fill="red",width=20)
ligne =canvas.create_line(600,0,600,600,fill="white",width=20)
ligne =canvas.create_line(600,0,0,0,fill="red",width=20)
ligne =canvas.create_line(0,0,0,0,fill="white",width=20)
canvas.grid()
balle = creer_balle()
mouvement()
racine.mainloop()
```
